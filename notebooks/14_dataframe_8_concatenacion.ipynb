```python
# concatenacion por filas
DF1 = pd.DataFrame(
    {
        "NOMBRE":["Jose","Maria"],
        "CARRERA":["Economia","Arquitectura"],
        "EDAD":[23,26]
    }
).set_index("NOMBRE")
DF1
```

```python
# para concatenar, ambos DF deben tener las mismas columnas
DF2 = pd.DataFrame(
    {
        "NOMBRE":["Aurora","Maria"],
        "CARRERA":["Medicina","Informatica"],
        "EDAD":[22,28]
    }
).set_index("NOMBRE")
DF2
```

```python
# combinar los DFs
DF = pd.concat([DF1, DF2])
```

```python
DF
```

```python
# concatenacion por columnas, LOS DFs deben tener la misma cantidad de filas ... Punto de fusi√≥n AUTOS por eso: .set_index("AUTOS")
DF1 = pd.DataFrame(
    {
        "AUTOS":["Nissan","Ford", "Audi"],
        "COLOR":["blanco","azul", "rojo"]
    }
).set_index("AUTOS")
DF1
```

```python
DF2 = pd.DataFrame(
    {
        "AUTOS":["Nissan","Ford", "Audi"],
        "MODELO":[2018,2020,2022]
    }
).set_index("AUTOS")
DF2
```

```python
# axis 1 se usa para realizar una combinacion por filas, 0 es por columna
DF = pd.concat([DF1, DF2], axis=1)
DF
```
